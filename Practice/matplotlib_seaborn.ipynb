```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv("asset/data.csv")
```

```python
df.head()
```

## Simple Line Plot (Matplotlib)

```python
plt.plot(df['potential_V'], df['x_position'])
plt.xlabel('Potential_V (V)')
plt.ylabel('x_position')
plt.title('Cyclic Voltammetry of WOâ‚ƒ Film')
plt.show()
```

## Scatter Plot

```python
plt.scatter(df['potential_V'], df['x_position'])
plt.xlabel('Potential_V (V)')
plt.ylabel('x_position')
```

## Multiple Curves on One Plot

```python
df1 = pd.read_csv("asset/x2.csv")
df2 = pd.read_csv("asset/x3.csv")
```

```python
plt.plot(df1['x'], df1['y'], label='x2')
plt.plot(df2['x'], df2['y'], label='x3')
plt.legend()
```

## Save Figures

```python
plt.tight_layout()
plt.savefig('asset/x2 and x3.png', dpi=300)
```

## Seaborn Basics

```python
sns.lineplot(data=df, x='potential_V', y='y_position')
#sns.scatterplot(data=df, x='composition_ratio', y='x_position')
```

```python
sns.scatterplot(data=df, x='composition_ratio', y='x_position')
```

## Histogram / Distribution

```python
sns.histplot(df['potential_V'], bins=30)
```

## Heatmap (Very Relevant for Mapping!) 

```python
pivot = df.pivot(index='y_position', columns='x_position', values='potential_V')
sns.heatmap(pivot, cmap='viridis')
```

## boxplot

```python
sns.boxplot(data=df, x='potential_V', y='y_position')
```
