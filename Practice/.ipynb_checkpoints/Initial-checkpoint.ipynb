## map

we can use map to apply a function or lambda to a stream of data (list, dict,..e.t.c). we can only use functions or lambdas which rturn data.

```python
#lets define a function
def multy3(nump):
    return(nump*3)
#lets define a list
base_list = [2, 4, 6, 8 ]
#now I want to apply multy3 function for each element of the list and create a new list
out_list = list(map(multy3, base_list))

print(out_list)
```

```python
list_name = ['Nipun', 'Nipuni', 'Shakila' ]
def makeGood(name):
    return('Good ' + name)
good_list = list(map(makeGood, list_name))
```

```python
good_list
```

# lamda

```python
goodie = lambda name: "good "+ name
# lets apply this lambda to the list name and get output

out_list_2 = list(map(goodie, list_name))
out_list_2
```

```python
# in this case im not seperately wroiting lambda, im using custom lambda

out_list_3 = list(map(lambda name: "Bad " + name, list_name))
```

```python
out_list_3
```

```python
# i need to multiply following list by 4 and save it as a list
b_list = [2, 6, 1]

out_list_4 = list(map(lambda num: num*4, b_list))
out_list_4
```

## filter

```python
c_filter = [2, 5, 10]
out_list_10 = list(filter(lambda num: num % 2 == 1, c_filter))
out_list_10
```

## tuple unpacking

```python
t = [(1, 2), (8, 1), (2, 8)]
```

```python
for item in t:
    print(item)
```

```python
for (a,b) in t:
    print(a)
    print(b)
```

```python
td = [(1, 2, 3), (7, 8, 9), (3, 4, 5)]
for item in td:
    print(item)
    
```

```python
for (a,b,c) in td:
    print(a)
    print(b)
    print(c)
```
